```python
# Q2. Missing & Duplicate Values

# Check missing values in each column
print("Missing Values in Each Column:")
print(df.isnull().sum())

# Check total missing values
print("\nTotal Missing Values:")
print(df.isnull().sum().sum())

# Check duplicate rows
duplicate_rows = df.duplicated().sum()

print("\nNumber of Duplicate Rows:")
print(duplicate_rows)

# Remove duplicate rows
df = df.drop_duplicates()

# Verify duplicates after removal
print("\nDuplicate Rows After Removal:")
print(df.duplicated().sum())

# Display final shape
print("\nShape After Removing Duplicates:")
print(df.shape)
```

```python
# Q4. Histogram of Numeric Features

import matplotlib.pyplot as plt
import seaborn as sns

# List of numeric columns
numeric_columns = ['price', 'mileage', 'year', 'engineSize', 'mpg']

# Create histogram for each numeric column
for col in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Q5. Count Plots of Categorical Features

import matplotlib.pyplot as plt
import seaborn as sns

# Find all categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(categorical_columns)

# Create count plots for each categorical column
for col in categorical_columns:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Q10. Complete Preprocessing Pipeline

import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load Dataset
df = pd.read_csv("ford_car_dataset - ford_car_dataset.csv")

# Remove Duplicates
df.drop_duplicates(inplace=True)

# Handle Missing Values
df.dropna(inplace=True)

# -------- EDA --------

# Histogram
numeric_columns = ['price','mileage','year','engineSize','mpg']

for col in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()

# Count Plots
categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=col,
                  order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.show()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(),
            annot=True,
            cmap='coolwarm')
plt.show()

# Input and Output Features
X = df.drop('price', axis=1)
y = df['price']

# One-Hot Encoding
X = pd.get_dummies(X, drop_first=True)

# Feature Scaling
scaler = StandardScaler()

numeric_cols = X.select_dtypes(include=['int64','float64']).columns
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

print("Final Shape:", X.shape)
print("\nFirst 5 Rows:")
print(X.head())
```
