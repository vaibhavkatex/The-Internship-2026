```python
df = pd.read_csv("ford_car_dataset_cleaned.csv")
df
```

```python

# Independent Features
X = df.drop("price", axis=1)

# Dependent Feature
Y = df["price"]

# Print shapes
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python

# Identify categorical columns
categorical_columns = X.select_dtypes(include="object").columns

print("Categorical Columns:")
print(categorical_columns)

# One-Hot Encoding
X_encoded = pd.get_dummies(X)

# Convert True/False to 0/1
X_encoded = X_encoded.astype(int)

# Show first 5 rows
print(X_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Numerical columns
numerical_columns = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# Create StandardScaler object
scaler = StandardScaler()

# Apply scaling
X_encoded[numerical_columns] = scaler.fit_transform(X_encoded[numerical_columns])

# Show first 5 rows
print(X_encoded[numerical_columns].head())
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    Y,
    test_size=0.33,
    random_state=42
)

# Print shapes
print("X_train Shape:", X_train.shape)
print("X_test Shape :", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape :", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Create Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Print intercept
print("Intercept:", model.intercept_)

# Print coefficients
print("Coefficients:")
print(model.coef_)
```

```python
# Make predictions
y_pred = model.predict(X_test)

# Print first 10 predictions
print("First 10 Predicted Values:")
print(y_pred[:10])

# Print first 10 actual values
print("\nFirst 10 Actual Values:")
print(y_test.iloc[:10])
```

```python
from sklearn.metrics import r2_score

# Calculate R² Score
r2 = r2_score(y_test, y_pred)

# Print R² Score
print("R² Score:", r2)
```

```python
import joblib

# Save Linear Regression model
joblib.dump(model, "LR_ford_car.pkl")

# Save StandardScaler
joblib.dump(scaler, "scaler.pkl")

# Save encoded column names
joblib.dump(X_encoded.columns.tolist(), "columns.pkl")

print("All files saved successfully!")
```
