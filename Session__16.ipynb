**Q5. Create distribution plots for all numerical columns (age, bmi, children,
charges):
Use sns.histplot() or sns.distplot()
Create one plot per column (total 4 plots)
Add meaningful titles and labels***

```python
import matplotlib.pyplot as plt
num_cols = ['age', 'bmi', 'children', 'expenses']

for col in num_cols:

    sns.histplot(df[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

**Q6. Create count plots for all categorical columns (sex, smoker, region):
Use sns.countplot() for each column
Add titles and labels for clarity**

```python
# List of categorical columns
cat_cols = ['sex', 'smoker', 'region']

# Create one count plot for each categorical column
for col in cat_cols:
    sns.countplot(x=df[col])
    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

**Q7. Create a heatmap to show the correlation between numerical variables.
Use sns.heatmap()
Annotate the values and use a suitable color palette**

```python
# Correlation matrix
correlation = df.corr(numeric_only=True)

# Heatmap
sns.heatmap(correlation,
            annot=True,      # Values show karega
            cmap='coolwarm', # Color palette
            fmt=".2f")       # 2 decimal places

plt.title("Correlation Heatmap")
plt.show()
```

**Q8. Perform the following analysis on the charges column:
Find the average insurance charges

Find the maximum and minimum charges

Group by smoker and compare the average charges for smokers vs non-
smokers**

```python
average_charges = df['expenses'].mean()
print("Average Insurance Charges:", average_charges)
```

```python
max_charges = df['expenses'].max()
min_charges = df['expenses'].min()

print("Maximum Charges:", max_charges)
print("Minimum Charges:", min_charges)
```

```python
smoker_avg = df.groupby('smoker')['expenses'].mean()

print(smoker_avg)
```

**Q9. Create a boxplot to show the distribution of charges with respect to smoker
and sex.
(You can create two separate boxplots or use hue parameter)**

```python
#Charges vs Smoker
sns.boxplot(x='smoker', y='expenses', data=df)

plt.title("Charges Distribution by Smoker")
plt.xlabel("Smoker")
plt.ylabel("Charges")

plt.show()
```

```python
# Charges vs Sex

sns.boxplot(x='sex', y='expenses', data=df)

plt.title("Charges Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Charges")

plt.show()
```
