*Q10. (Mini Project - NumPy Application)
Create a Simple Statistics Calculator using NumPy:
Write a program that:
Asks the user how many numbers they want to generate.
Generates that many random numbers between 10 and 100 using
np.random.randint().
Stores them in a 1D array.
Prints:
The array
Mean, Median, Standard Deviation
Minimum and Maximum
Reshapes into a 2D array (if possible) and prints row-wise sum.
Use proper comments and exception handling where needed.*

```python
import numpy as np

try:
    # Ask the user how many numbers to generate
    n = int(input("How many random numbers do you want to generate? : "))

    if n <= 0:
        print("Please enter a number greater than 0.")

    else:
        # Generate random numbers between 10 and 100
        arr = np.random.randint(10, 101, n)

        # Print the array
        print("\nGenerated Array:")
        print(arr)

        # Statistics
        print("\n----- Statistics -----")
        print("Mean :", np.mean(arr))
        print("Median :", np.median(arr))
        print("Standard Deviation :", np.std(arr))
        print("Minimum :", np.min(arr))
        print("Maximum :", np.max(arr))

        # Reshape into 2D array if possible
        print("\n----- 2D Matrix -----")

        rows = int(np.sqrt(n))

        if rows > 1 and n % rows == 0:
            cols = n // rows

            matrix = arr.reshape(rows, cols)

            print(matrix)

            print("\nRow-wise Sum:")
            print(np.sum(matrix, axis=1))

        else:
            print("Cannot reshape into a proper 2D matrix.")

except ValueError:
    print("Invalid Input! Please enter an integer.")
```
