```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import joblib
```

```python
# Load the dataset
df = pd.read_csv("./ford_car_dataset - ford_car_dataset.csv")
df.shape
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Check for year outliers
print("Years in dataset:", sorted(df['year'].unique()))

# Remove the 2060 year outlier (typo in dataset)
df = df[df['year'] <= 2021]
print(f"Dataset Shape after cleaning: {df.shape}")
```

```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
df.describe()
```

```python
# Apply One-Hot Encoding to categorical columns
X_encoded = pd.get_dummies(X, columns=['model', 'transmission', 'fuelType'])

# Save the list of encoded columns for future alignment
encoded_columns = X_encoded.columns.tolist()
joblib.dump(encoded_columns, 'columns.pkl')

print(f"Total features after one-hot encoding: {len(encoded_columns)}")
```

```python
# Fit standard scaler only on the numerical features
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

scaler = StandardScaler()
X_encoded[numerical_cols] = scaler.fit_transform(X_encoded[numerical_cols])

# Save the scaler object
joblib.dump(scaler, 'scaler.pkl')
print("Scaler successfully saved!")
```

```python
# Split the data for evaluation
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize and train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate performance
train_r2 = model.score(X_train, y_train)
test_r2 = model.score(X_test, y_test)
print(f"Train R2 Score: {train_r2:.4f}")
print(f"Test R2 Score: {test_r2:.4f}")
```

```python
# Retrain model on full dataset
final_model = LinearRegression()
final_model.fit(X_encoded, y)

# Save the trained model
joblib.dump(final_model, 'LR_model.pkl')
print("Final model successfully trained and saved!")
```
